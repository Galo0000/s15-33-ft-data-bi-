```python
import pandas as pd
```

```python
df_clubs = pd.read_json('./Data/liga brasil/clubs.json')
df_clubs.head()
```

```python
# Extraer el número y guardarlo en una nueva columna llamada 'club_id'
df_clubs['club_id'] = df_clubs['href'].str.extract(r'/(\d+)$')

# Eliminar la columna original
df_clubs.drop(columns=['href'], inplace=True)

df_clubs.head()
```

Se elimina la columna coach name por se totalmente nula

```python
df_clubs.drop('coach_name',axis=1,inplace=True)
df_clubs.head()
```

```python
for a in range(len(df_clubs)):
    print(df_clubs['parent'][a])
```

Se elimina la columna parent, por ser redundante

```python
df_clubs.drop('parent',axis=1,inplace=True)
df_clubs.head()
```

```python
df_clubs['type'].unique()
```

Se elimina la columna type por ser redundante

```python
df_clubs.drop('type',axis=1,inplace=True)
df_clubs.head()
```

```python
df_clubs.info()
```

```python
#Eliminar el signo '%', los espacios vacíos, convierte a valor flotante y por ultimo divide el valor por 100
df_clubs['foreigners_percentage'] = df_clubs['foreigners_percentage'].str.replace('%', '').astype(float) / 100
df_clubs.head()
```

```python
#Elimina la palabra 'Seats', los "." y se transofrma a entero
df_clubs['stadium_seats'] = df_clubs['stadium_seats'].str.replace('Seats', '').str.replace('.', '').astype(int)
df_clubs.head()
```

```python
#Elimina la palabra simbolos, letras y transoforma a valor flotante
df_clubs['net_transfer_record'] = df_clubs['net_transfer_record'].str.replace('m', '').str.replace('€-', '').str.replace('+€', '').str.replace('k', '').str.replace('+-', '').astype(float)
df_clubs.head()
```

```python
df_clubs.to_csv('./Data/liga brasil/clubs.csv',index=False)
```
